```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('../data/application_train.csv')

# Overview
print(f"Shape: {df.shape}")
df.head()
```

```python
# Check missing values
missing = df.isnull().sum()
print("NAs: ",df.isnull().sum().sum())
missing = missing[missing > 0].sort_values(ascending=False)
missing_percentage = (missing / df.shape[0]) * 100 #percentage of missing data
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_percentage}) #combine into a new df

missing_df
```

```python
# droping columns with alot of missing data
df = df.drop(columns=missing_df[missing_df['Missing %'] > 40].index)

# Impute remaining missing values
df.fillna(df.median(numeric_only=True), inplace=True)
print(df.select_dtypes(include='number').isna().sum().sum())
df.shape
```

## Dropping Statistically unnecessary columns

```python
corr_matrix = df.corr(numeric_only=True).abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_cols = [column for column in upper.columns if any(upper[column] > 0.95)]
print("Highly correlated columns to consider dropping:", high_corr_cols)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop(columns=['TARGET']).select_dtypes(include='number').fillna(0)
y = df['TARGET']

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

importances = pd.Series(rf.feature_importances_, index=X.columns)
low_importance_cols = importances[importances < 0.005].index
print("Low importance columns:", low_importance_cols.tolist())
```

```python
from sklearn.feature_selection import mutual_info_classif

X = df.drop(columns='TARGET')
y = df['TARGET']
mi = mutual_info_classif(X.select_dtypes(include='number'), y)
mi_series = pd.Series(mi, index=X.select_dtypes(include='number').columns)
low_mi_cols = mi_series[mi_series < 0.01].index
print("Low mutual information columns:", low_mi_cols.tolist())
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.0)
selector.fit(df.select_dtypes(include='number'))
low_variance_cols = df.select_dtypes(include='number').columns[~selector.get_support()]
print("Low variance columns:", low_variance_cols.tolist())
```

```python
cols_to_drop = set(list(low_variance_cols) + list(low_mi_cols) + list(low_importance_cols) + list(high_corr_cols))

df.drop(columns=cols_to_drop, inplace=True)
```
